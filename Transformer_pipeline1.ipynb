```python
from transformers import pipeline
```

```python
classifier =pipeline("sentiment-analysis")
classifier("I have been waiting to learn from HuggingFace for quite sometime.")
```

```python
classifier("I'm hate waiting for GF, cant wait her smile!!"," Her smile makes my day")
```

```python
classifier("I'm so happy")
```

```python
classifier2= pipeline("zero-shot-classification")
```

```python
classifier2("This is  a course about transformers library from huggingface", candidate_labels=["education", "politics", "business","intemacy"],)
```

```python
generator =pipeline("text-generation")

generator("In this course, we will teach you how to")
```

```python
generator2  = pipeline("text-generation", model='distilgpt2')
generator2("In this couse, we will teach you how to", max_length =100, num_return_sequences=5)
```

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> model", top_k=2)
```

```python
unmasker("I wanna <mask> your <mask>", top_k=2)
```

```python
ner = pipeline("ner", grouped_entities= True)
```

```python
preds= ner("my name is Sharath")
```

```python
print(*preds, sep="\n")
```

```python
preds = [
    {
        "entity": pred["entity_group"],
        "score": round(pred["score"], 4),
        "word": pred["word"],
        "start": pred["start"],
        "end": pred["end"],
    }
    for pred in preds
]
```

```python
print(*preds, sep="\n")
```
