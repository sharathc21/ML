```python
raw_datasets = load_dataset('glue','mrpc')
```

```python
checkpoint ='bert-base-cased'
```

```python
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
def tokenize_func(ex):
  return tokenizer(ex['sentence1'], ex['sentence2'], truncation=True)
```

```python
tokenized_data = raw_datasets.map(tokenize_func, batched=True)
```

```python
tokenized_data['train'][:5]
```

```python
tokenized_data['train'].features
```

```python
tokenized_data = tokenized_data.remove_columns(['idx', 'sentence1', 'sentence2'])
```

```python
tokenized_data =tokenized_data.with_format('torch')
```

```python
tokenized_data['train'].features
```

```python
from torch.utils.data import DataLoader
```

```python
from transformers import DataCollatorWithPadding
```

```python
data_collator = DataCollatorWithPadding(tokenizer)
```

```python
train_dataloader = DataLoader(tokenized_data['train'], batch_size=16, shuffle=True, collate_fn=data_collator)
```

```python
for step, batch in enumerate(train_dataloader):
  print(batch["input_ids"].shape)
  if step>5:
    break
```

Trainer API


```python
from transformers import AutoModelForSequenceClassification
```

```python
model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)
```

```python
from transformers import TrainingArguments
```

```python
training_args1 = TrainingArguments('test-trainer', per_device_eval_batch_size=16, per_device_train_batch_size=16, num_train_epochs=5, learning_rate=2e-5, weight_decay = 0.01)
```

```python
from transformers import Trainer
```

```python
trainer = Trainer (model, training_args1, train_dataset=tokenized_data['train'], eval_dataset=tokenized_data['validation'], data_collator=data_collator, tokenizer = tokenizer)
```

```python
trainer.train()
```

```python
predictions = trainer.predict(tokenized_data['validation'])
print(predictions.predictions.shape, predictions.label_ids.shape)
```

```python
import numpy as np
```

```python
from datasets.load import load_metric
metric =load_metric('glue', 'mrpc')
```

```python
preds = np.argmax(predictions.predictions, axis=-1)
```

```python
metric.compute(predictions=preds, references=predictions.label_ids)
```

```python
def compute_metrics(eval_preds):
  logits, labels =eval_preds
  predictions =np.argmax(logits, axis=-1)
  return metric.compute(predictions=predictions, references = labels)
```

```python
training_args2 =TrainingArguments('test-trainer', evaluation_strategy ='epoch')
```

```python
trainer = Trainer (model, training_args1, train_dataset=tokenized_data['train'], eval_dataset=tokenized_data['validation'], data_collator=data_collator, tokenizer = tokenizer, compute_metrics=compute_metrics)
```

```python
trainer.train()
```
